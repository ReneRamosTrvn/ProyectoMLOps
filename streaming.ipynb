```python
import pandas as pd
from datetime import datetime
```

```python
amzdf = pd.read_csv('./datasets/amazon_prime_titles.csv')
amzdf.head()
```

```python
amzdf['show_id'] = amzdf['show_id'].apply(lambda x: 'a' + x)
amzdf.set_index('show_id', inplace=True)
```

```python
amzdf.head(3)
```

```python
amzdf['rating'].fillna('G', inplace=True)
amzdf.head(8)
```

```python
amzdf.info()
```

```python
amzdf['date_added'] = pd.to_datetime(amzdf.date_added)
```

```python
amzdf.head()
```

```python
amzdf = amzdf.applymap(lambda s: s.lower() if type(s) == str else s)
amzdf.head()
```

```python
amzdf[['duration_int', 'duration_type']] = amzdf['duration'].str.split(' ', 1, expand = True)
amzdf.head()
```

```python
amzdf.drop(columns=['duration'], inplace=True)
amzdf.head()
```

```python
amzdf['duration_int'] = amzdf['duration_int'].astype(int) 
amzdf.info()
```
